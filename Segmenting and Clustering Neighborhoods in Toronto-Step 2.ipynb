```python
import pandas as pd
import numpy as np
```

```python
#Getting data; The dataframe with three columns: PostalCode, Borough, and Neighborhood
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
wiki_data=pd.DataFrame(df[0])
wiki_data
```

```python
# Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

df1= wiki_data[wiki_data['Borough'] != 'Not assigned']
df1.head(30)
```

```python
#In the last cell of your notebook, use the .shape method to print the number of rows of your dataframe.

df1.shape
```

### Step 2- Reading Geographical Co-ordinates

```python
geo_location= pd.read_csv('http://cocl.us/Geospatial_data')
geo_location
df1['Latitude']= geo_location['Latitude']
df1['Longitude']=geo_location['Longitude']
df1
```
